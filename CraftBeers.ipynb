```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
brew = pd.read_csv('breweries.csv')
beer = pd.read_csv('beers.csv')
```

```python
brew.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
beer.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
plot = brew.groupby('state')['name'].count().nlargest(20).plot(kind='bar', \
               title='States with the Most Breweries', \
               colormap='summer',  )
plot.set_ylabel('Number of Breweries')
```

```python
plot2 = brew.groupby('city')['name'].count().nlargest(15).plot(kind='bar', \
               title='Cities with the Most Breweries', \
               colormap='summer',  )
plot2.set_ylabel('Number of Breweries')
```

```python
plot3 = beer.groupby('style')['style'].count().nlargest(15).plot(kind='bar', \
               title='Popular Styles of Beer', \
               colormap='summer',  )
plot3.set_ylabel('Styles of Beers')
```

```python
plot4 = beer.groupby('style')['abv'].mean().nlargest(15).plot(kind='bar', \
               title='Beer Styles with Highest Average Alcohol by Volume', \
               colormap='summer' )
```
